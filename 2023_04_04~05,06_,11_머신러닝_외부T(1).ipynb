```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup


def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub = row['출판사']
  year = row['발행년도']
  
  isbn = row['ISBN']
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}' 

  r = requests.get(url)

  soup = BeautifulSoup(r.text, 'html.parser')

  try:
    if pd.isna(title):
      title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  except AttributeError:
    pass

  try:
    if pd.isna(author):
      authors = soup.find('span', attrs={'class':'info_auth'}).find_all('a') 
      author = list(tag.get_text() for tag in authors) 
  except AttributeError:
    pass

  try:
    if pd.isna(pub):
      pub = soup.find('span', attrs={'class':'info_pub'}).find('a').get_text()
  except AttributeError:
    pass
  
  try:
    if pd.isna(year):
      year_str = soup.find('span', attrs={'class':'info_date'}).get_text()
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass

  try:
    if pd.isna(year):
      year_str = soup.find('span', attrs={'class':'info_date'}).get_text()
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass

  try:
    if pd.isna(year):
      year_str = soup.find('span', attrs={'class':'info_date'}).get_text()
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass
  
  return title,author,pub,year

updated_sample = ns_book5[na_rows].head(10).apply(get_book_info, axis=1, result_type='expand')
updated_sample
```

```python
import gdown 
import pandas as pd


#gdown 패키지 이용, 서울 남산도서관 데이터 다운
import gdown
gdown.download('http://bit.ly/3RhoNho', 'ns_202104.csv', quiet=False); 

def data_cleaning(failName):

  #파일을 데이터프레임으로 읽기
  ns_df = pd.read_csv('{}'.format(failName),  low_memory=False, encoding='utf-8')

  #NaN 하나 이상인 열을 삭제
  ns_book = ns_df.dropna(axis=1, how='all')

  #대출건수를 합치기 위해 필요한 행만 추출하여 새로운 데이터 프레임만들기 ('도서명','저자','ISBN','권','대출건수')
  count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]

  #'도서명','저자','ISBN','권'을 기준으로 대출건수를 groupby하기 
  loan_count = ns_book.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()


  #원본 데이터프레임에서 중복된 행을 제외하고 고유한 행만 추출하여 복사
  dup_rows = ns_df.duplicated(subset=['도서명','저자','ISBN','권'], keep=False)
  unique_rows = ~dup_rows
  ns_book3 = ns_book[unique_rows].copy()

  #도서명, 저자, ISBN, 권을 인덱스로 설정
  ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)

  #loan_count에서 저장된 누적 대출건수 업데이트
  ns_book3.update(loan_count)

  #인덱스 설정
  ns_book4= ns_book3.reset_index()
  
  #원본 데이터프레임의 열 순서 변경
  ns_book4 = ns_book4[ns_book.columns]
  return ns_book4

ns_book4 = data_cleaning('ns_202104.csv')
ns_book4.head(5)
```
