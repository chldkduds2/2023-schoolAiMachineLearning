```python
import pandas as pd
df1 = {'번호' : ['1','2','3', '4', '5'] ,
        '이름' : ['개나리','진달래','무궁화', '모과', '월계수'],
        '생년' : [1975 , 1980 , 1992, 2004, 1998] }
stu1=pd.DataFrame(df1)
stu1
```

```python
df2 = {'번호' : ['1','2', '4', '6'] ,
        '이름' : ['개나리','진달래','물망초', '모과'],
        '출생지' : ['한국' , '한국', '스위스', '캐나다'] }
stu2=pd.DataFrame(df2)
stu2
```

```python
stu3 = stu1.merge(stu2)
stu3
```

```python
pd.merge(stu1,stu2)
```

```python
stu3 = stu1.merge(stu2)
stu3
```

```python
stu3 = pd.merge(stu1, stu2, left_on='이름', right_on='이름', how='left')
stu3
```

```python
stu3 = pd.merge(stu1, stu2, left_on='이름', right_on='이름', how='outer')
stu3
```

```python
pd.merge(stu1, stu2,  how='left')
```

```python
pd.merge(stu1, stu2,  how='right')
```

```python
pd.concat([stu1, stu2], axis=1)
```

```python
stu3= pd.concat([stu1, stu2], axis=0)
stu3
```

```python
stu3= pd.concat([stu1, stu2], axis=0)
stu3['나이']=2023-stu3['생년']


stu3.loc[(stu3['나이'] >= 8) & (stu3['나이'] <= 13) , '구분'] = '초'
stu3.loc[(stu3['나이'] >= 14) & (stu3['나이'] <= 16) , '구분'] = '중'
stu3.loc[(stu3['나이'] >= 17) & (stu3['나이'] <= 19) , '구분'] = '고'
stu3.loc[stu3['나이'] >= 20  , '구분'] = '성인'
stu3.loc[stu3['나이'] <=8, '구분'] = '미취학'

stu3



  
```
