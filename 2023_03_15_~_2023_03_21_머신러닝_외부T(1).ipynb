# API 사용하기
- 파이썬에서 json 데이터 다루기

```python
#json 형식으로 파이썬 딕셔너리 생성
d = {'name': '혼자 공부하는 데이터 분석'}
print(d['name'])
```

```python
#파이썬 객체를 json 문자열로 변환하기 (json.dumps()함수)
import json
d_str = json.dumps(d, ensure_ascii=False)
print(d_str)
```

```python
print(type(d_str)) #class -> json 형태
```

```python
#json 파일을 딕셔너리로 변환하기 (json.loads()함수)
d2=json.loads(d_str)
print(d2['name'])
print(type(d2))
print(d2)
```

```python
d3=json.loads('{"name": "혼자 공부하는 데이터 분석", "author":"박해선","year":2022}')
print(d3['name'])
print(d3['author'])
print(d3['year'])
```

```python
#딕셔너리 안에 배열 접근
d3=json.loads('{"name": "혼자 공부하는 데이터 분석", "author":["박해선","홍길동"],"year":2022}') # json의 하나의 객체로
print(d3['author'][1])
```

```python
# json.loads()함수를 사용해 파이썬 객체 변환
d4_str="""
[
  {"name":"혼자 공부하는 데이터 분석", "author":"박해선", "year":2022},
  {"name":"혼자 공부하는 머신러닝+딥러닝", "author":"박해선", "year":2020}
]
"""
d4=json.loads(d4_str)
print(d4[0]["name"])
print(d4[1]["name"])
```

```python
#josn문자열을 데이터 프레임으로 변환하기 (read_json()함수)
import pandas as pd
pd.read_json(d4_str)
```

```python
#파이썬 객체로 만든 Json문자열을 데이터 프레임으로 만들기
pd.DataFrame(d4)
```

- 파이썬에서 XML(웹 개발 문법) 다루기

```python
x_str="""
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""
```

```python
#문자열을 XML형식으로 변환
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
print(type(book))
print(book.tag) #해당 변수의 XML문법 적용 태그 출력
```

```python
#자식 엘리먼트 확인하기 (findtext()메서드)
book_childs=list(book)
print(book_childs)
```

```python
#book_childs리스트 각 항목을 변수에 할당하고 text속성으로 엘리먼트 텍스트 출력

#잘 못된 예시 (안정성 X)
name, author,year=book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
#잘 된 예시 (안정성 O)
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')
print(name)
print(author)
print(year)
```

```python
x2_str="""
 <books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>

    <book>
    <name>혼자 공부하는 머신러닝+딥러닝</name>
    <author>홍길동</author>
    <year>2020</year>
  </book>
 </books>
 """
```

```python
#부모 엘리먼트 확인 (fromstring())
import xml.etree.ElementTree as et
books = et.fromstring(x2_str)
print(books.tag)
```

```python
#여러 개의 자식 엘리먼트 확인 (findall())
for book in books.findall('book'):
  name=book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  print(name)
  print(author)
  print(year)
  print('-'*30)
```

```python
#XML을 josn으로 판단스를 이용해 바꾸기
pd.read_xml(x2_str)
```

```python
#파이썬으로 API 호출하기 (requests 패키지)
import requests
url = "http://data4library.kr/api/loanItemSrch?format=json&startDt=2021-04-01&endDt=2021-04-03&age=2-&authKey=90c0ffdea31504bad9e1b08ab4580b76e5f20b3c2b74914a426e87208779d9a3"
r = requests.get(url)
```

```python
#json 파일로 불러온 데이터 변형
data = r.json()
print(data)
```

```python
data
```

```python
#필요한 정보 빼내오기 (도서책이름)
books = []
for d in data['response']['docs']:
  books.append(d['doc'])
```

```python
books
```

```python
#판단스 DataFrame 클래스에 리스트 넘기기
import pandas as pd
books_df = pd.DataFrame(books)
books_df
```

```python
books_df.to_json('20s_best_book.json')
books_df.head()
```
