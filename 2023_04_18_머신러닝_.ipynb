```python
import csv
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt


plt.rc('font', family='NanumBarunGothic')


f = open('/content/1군전염병발병현황_년도별.csv', 'r', encoding='cp949')

rdr = csv.reader(f)


df = pd.DataFrame(rdr)


df.rename(mapper = df.iloc[0], axis=1, inplace = True)

df.drop([0], axis=0, inplace=True)


df['콜레라'] = df['콜레라'].apply(int)

df['장티푸스'] = df['장티푸스'].apply(int)

df['이질'] = df['이질'].apply(int)

df['대장균'] = df['대장균'].apply(int)


fig, ax = plt.subplots(figsize=(15,5))

bar_width = 0.15


year = np.arange(2002, 2013)


b1 = plt.bar(year-bar_width, df['콜레라'], bar_width, alpha=0.4, color='red', label='콜레라')

b2 = plt.bar(year, df['장티푸스'], bar_width, alpha=0.4, color='blue', label='장티푸스')

b3 = plt.bar(year + bar_width, df['이질'], bar_width, alpha=0.4, color='green', label='이질')

b4 = plt.bar(year + bar_width * 2, df['대장균'], bar_width, alpha=0.4, color='yellow', label='대장균')

plt.legend()
plt.xticks(year, df['년도'])

plt.show()
f.close()
```
