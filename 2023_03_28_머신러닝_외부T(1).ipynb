# 중복된 행 찾기

그 외의 중복되지 않은 나머지 모든 행은 False

```python
#sum 함수 : 중복된 행의 갯수를 셀 수 있음 (True=1)
sum(ns_book.duplicated())
#데이터 프레임의 행은 모두 고윳값을 가짐 따라서 -> 0
```

# 그룹별로 모으기

**groupby() 메소드**

```python
count_df =ns_book[['도서명', '저자', 'ISBN', '권','대출건수']]
count_df.head()
```

```python
group_df = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False) #dropna : 결측값을 계산에서 제외할지 여부(NaN 포함 열 제거)
loan_count = group_df.sum()
group_df.head()
```

```python
loan_count
```

```python
#한 줄
loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
loan_count.head()
```

**~ 연산자**
- bitwise 연산자
- 이진수 1 보완
- 불리언 값 반대로

**copy 메소드**
- 얕은 사본만들기

```python
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권']) #중복된 행을 True
unique_rows = ~dup_rows #불리언 배열을 반전시키고 고유한 행을 True 
ns_book3= ns_book[unique_rows].copy()#고유한 행만 선택
```

```python
#중복된 행이 없는지 확인
sum(ns_book3.duplicated(subset=['도서명','저자','ISBN','권']))
```
