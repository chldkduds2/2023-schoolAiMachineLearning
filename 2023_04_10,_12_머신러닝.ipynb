```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')


df = pd.read_csv('학생별과목별성적_3명.csv', encoding='CP949')
df = df.loc[((df['과목']=='수학') | (df['과목']=='영어') | (df['과목']=='국어')) & (df['점수']>=80)]

df1 = df.loc[(df['이름']=='박지영')]
df2= df.loc[(df['이름'] == '김태함')]
df3 = df.loc[(df['이름']=='권혁진')]

plt.style.use('ggplot')
fig = plt.figure(figsize=(10,10))
fig.set_facecolor('white')
ax = fig.add_subplot()



ax.plot(df1['이름'],df1['점수'], 'o',label='박지영', color='red')
ax.plot(df2['이름'],df2['점수'], 'o',label='김태함', color='blue')
ax.plot(df3['이름'],df3['점수'], 'o',label='권혁진', color='purple')

plt.title('학생별 과목 성적')
plt.xlabel('학생명')
plt.ylabel('점수 (단위 : 점)')


plt.legend(loc='upper left')
plt.show()
```
