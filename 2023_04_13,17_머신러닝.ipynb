```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig=plt.figure()
ax=fig.add_subplot(1,1,1)   
labels=['eng','math']
x=[3,4]
y=[1,2]
colors=['red', 'blue']
ax.bar( x,y, tick_label=labels,color=colors)

for i,v in enumerate(y):
  ax.text(v+2, i+1.1, str(v),fontsize=9,color='black',fontweight='bold',ha='center', va='bottom',)

plt.show()
```

```python
import pandas as pd

df = pd.read_csv('1군전염병발병현황_년도별.csv', encoding='cp949')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
fig=plt.figure()
ax=fig.add_subplot()  
bar_width = 3

index = np.arange(df['년도'].count())
labels = df['년도']


b1 = plt.bar(index, df['콜레라'], bar_width, color="darkblue", label=df['년도'])
b2 = plt.bar(index, df['장티푸스'], bar_width, color="red", label=labels)
b3 = plt.bar(index, df['이질'], bar_width, color="green", label=labels)
b4 = plt.bar(index, df['대장균'], bar_width, color="blue", label=labels)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rc('font', family='NanumBarunGothic')

plt.style.use('ggplot')
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot()  
bar_width = 0.25

nrow = df.shape[0]
index = np.arange(nrow)


b1 = plt.bar(index-bar_width, df['콜레라'], bar_width, color="red", label='콜레라')
b2 = plt.bar(index, df['장티푸스'], bar_width, color="blue", label='장티푸스')
b3 = plt.bar(index+bar_width, df['이질'], bar_width, color="purple", label='이질')
b4 = plt.bar(index+bar_width*2, df['대장균'], bar_width, color="black", label='대장균')

plt.legend()
plt.xticks(index, df['년도'])
plt.ylabel('발생 건수')




plt.show()
```
